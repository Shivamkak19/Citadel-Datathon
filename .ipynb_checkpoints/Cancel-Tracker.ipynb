```python
import numpy as np
import pandas as pd
```

```python
traffic = pd.read_csv("../../Summer_Invitational_2023_Datathon_Datasets/flight_traffic.csv")
fares = pd.read_csv("../../Summer_Invitational_2023_Datathon_Datasets/fares.csv")
```

```python
traffic.head()
```

```python
delays = traffic[['air_system_delay', 'airline_delay', 'security_delay', 'aircraft_delay']]
delays.fillna(0, inplace= True)

# Average delay tied to non-weather conditions (human error)
avgDelay = delays.mean(axis=1)

print(type(avgDelay), type(delays))

print(avgDelay)

delays.head()
```

```python
traffic["avgDelay"] = avgDelay

traffic.head()
```

```python
# Columsn to slice by for total 
col_list = fares.columns[5:]
print(col_list)
```

```python
faresClean = fares.copy()
# Calculate the total revenue for columns 5 to 10 (inclusive)
faresClean['total_revenue'] = (fares.iloc[:, 5:] * fares.iloc[:, 5:].columns.astype(float)).sum(axis=1)

# Calculate the total number of tickets sold for columns 5 to 10 (inclusive)
faresClean['total_tickets'] = fares[col_list].sum(axis=1)

# Calculate the true average
faresClean['true_average'] = faresClean['total_revenue'] / faresClean['total_tickets']

faresClean.head(20)
```
