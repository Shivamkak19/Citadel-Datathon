```python
airline_id = airlines["airline_id"].tolist()
print(airline_id)
```

```python
# Create subset of traffic for cancelled flights
cancel = traffic[traffic["cancelled"] != 0]

cancel.tail(40)
```

```python
#Count how many cancellations each airline has

totals = cancel.groupby("airline_id").size()

print(totals)
```

```python
cancelById = []
for id in airline_id:
    df = cancel[cancel["airline_id"] == id].copy()
    cancelById.append(df)

cancelById
```
