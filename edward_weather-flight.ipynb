```python
import numpy as np
import pandas as pd
import openpyxl
```

```python
airline_list=pd.read_csv('airlines.csv')
airline_list
```

```python
df=pd.read_csv("flight_traffic.csv")
df[df['airline_id']=="AA"]
df
```

```python
airline_list['Departure Delay']=0
airline_list['Arrival Delay']=0
airline_list['Taxi Time']=0
airline_list['Proportion on Time']=0

for i,j in airline_list.iterrows():
    df1=df.loc[df['airline_id']==airline_list['airline_id'][i]].copy()
    df1['DepartureDelay']=df1['scheduled_departure']-df1['actual_departure']
    df1['ArrivalDelay']=df1['scheduled_arrival']-df1['actual_arrival']


    onTime=df1['airline_delay'].notnull().sum()

    airline_list['Proportion on Time'][i]=onTime/(len(df1['airline_delay']))

    airline_list['Arrival Delay'][i]=df1['ArrivalDelay'].mean(skipna=True)
    airline_list['Departure Delay'][i]=df1['DepartureDelay'].mean(skipna=True)
    airline_list['Taxi Time'][i]=df1['taxi_out'].mean(skipna=True)

airline_list.to_excel("output.xlsx")
```

```python
df['Date']=0
df['Date']=pd.to_datetime(df[['year','month','day']])
df
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(subset = ['actual_departure', 'actual_arrival'],inplace=True)
df['actual_departure'] = df['actual_departure'].astype('int')
df['actual_departure'] = df['actual_departure'].astype(str).str.zfill(4).astype(int)
df['actual_arrival'] = df['actual_arrival'].astype('int')
df['actual_arrival'] = df['actual_arrival'].astype(str).str.zfill(4).astype(int)

df['arrival_integer']=df['year']*100000000+df['month']*1000000+df['day']*10000+df['actual_arrival']
df['departure_integer']=df['year']*100000000+df['month']*1000000+df['day']*10000+df['actual_departure']
df['arrival_datetime']=pd.to_datetime(df['arrival_integer'], format='%Y%m%d%H%M%S')
df['departure_datetime']=pd.to_datetime(df['departure_integer'], format='%Y%m%d%H%M%S')
df
```
